#Exploratory data analysis

The idea is to use the House Prices dataset available in Kaggle tutorials as an example for machine learning for regression problems.

```python
import pandas as pd

df = pd.read_csv('/content/train.csv').drop(['Id'], axis=1)
df
```

```python
to_remove = df.isnull().sum().sort_values(ascending=False).head(16).index
to_remove
```

```python
df = df.drop(to_remove, axis=1).dropna()
df
```

```python
import numpy as np

numerical_features = df.select_dtypes(include=[np.number]).drop('SalePrice', axis=1).columns
categoricals_features = df.select_dtypes(exclude=[np.number]).columns

def normalize(s):
    #StandardScaler
    mean = s.mean()
    std = s.std()
    s = (s-mean)/std

    #MinMaxScaler
    max = s.max()
    min = s.min()
    s = (s-min)/(max-min)
    return s

for f in numerical_features:
    df[f] = normalize(df[f])
df = pd.get_dummies(df, columns=categoricals_features)

df
```

#The scikit-learn KNeighborsRegressor algorithm

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
random_state = 0

def apply_cv(df, model, random_state=0):
    X = df.drop('SalePrice', axis=1)
    y = df['SalePrice']

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)
    scores = cross_val_score(model, X, y, cv=cv, n_jobs=-1)
    return scores

apply_cv(df, KNeighborsRegressor(n_neighbors=5, weights='distance', algorithm='brute'))
```

#Implementation of the KNeighborsRegressor algorithm using the NumPy library in Python

```python
def get_X_y(df, target):
    X = df.drop(target, axis=1)
    y = df[target]
    return X, y

df_train = df.drop(df.tail(290).index)
X_train, y_train = get_X_y(df_train, 'SalePrice')

df_test = df.tail(290)
X_test, y_test = get_X_y(df_test, 'SalePrice')

class KNNRegressor:
    def __init__(self, X_train, y_train, k=5, p=1):
        self.X_train = X_train.values
        self.y_train = y_train.values[:, np.newaxis]
        self.k = k
        self.p = p

    def predict(self, X_test):
        X_train = self.X_train
        y_train = self.y_train
        X_test = X_test.values

        y_pred = []
        for i in range(len(X_test)):
            distance_list = []
            for j in range(len(X_train)):
                distance_list.append(self.get_distance(row_1=X_test[i], row_2=X_train[j]))
                
            distance_array = np.array(distance_list)[:, np.newaxis]
            self.distance_array = distance_array ###tirar depois
            y_pred.append(self.get_weighted_average(distance_array))

        self.y_pred = np.array(y_pred)
        return self.y_pred

    def get_distance(self, row_1, row_2):
        p = self.p        
        sum = ((np.abs(row_1 - row_2))**p).sum()
        return sum**(1/p)

    def get_weighted_average(self, distance_array):
        y_train = self.y_train
        k = self.k

        A = np.hstack((distance_array, y_train))
        A = A[A[:,0].argsort()]

        while A[k, 0] == A[k-1, 0]:
            k += 1
        A = A[:k]

        A[:,0] = self.get_inv_dist(A[:,0])

        if A[0, 0] == float('inf'):
            weighted_average = A[0, 1]
        else:
            B = A[:, 0] * A[:, 1]
            weighted_average = B.sum()/A[:,0].sum()
        
        return weighted_average

    def get_inv_dist(self, array):
        inv_list = []
        for i in array:
            if i == 0:
                inv_list.append(float('inf'))
            else:
                inv_list.append(1/i)
        return np.array(inv_list)

    def get_score(self, y_test):
        y_test = y_test.values
        if hasattr(self, 'y_pred') == False:
            print('No prediction has been made yet.')
            return

        y_pred = self.y_pred
        SSE = ((y_test - y_pred)**2).sum()
        SST = ((y_test - y_test.mean())**2).sum()
        score = 1 - (SSE/SST)
        return score

knn = KNNRegressor(X_train, y_train)
knn.predict(X_test)
knn.get_score(y_test)
```

```python
from math import ceil

class CrossValidation:
    def __init__(self, df, target, n=5):
        self.df = df
        self.target = target
        self.n = n

    def get_score_list(self):
        df = self.df.copy()
        n = self.n

        df = df.sample(frac=1)
        rows_per_split = ceil(len(df)/n)

        split_list = [df.iloc[i:i + rows_per_split, :] for i in range(0, len(df), rows_per_split)]
        self.split_list = split_list

        score_list = []
        for i in range(n):
            df_train = pd.concat(split_list[:i] + split_list[i+1:])
            df_test = split_list[i]

            X_train, y_train = self.get_X_y(df_train)
            X_test, y_test = self.get_X_y(df_test)

            knn = KNNRegressor(X_train, y_train)
            knn.predict(X_test)
            score_list.append(knn.get_score(y_test))

        self.score_list = score_list
        return score_list

    def get_X_y(self, df):
        target = self.target

        X = df.drop(target, axis=1)
        y = df[target]
        return X, y

CrossValidation(df, target='SalePrice').get_score_list()
```
